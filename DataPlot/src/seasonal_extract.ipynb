```python
def findDominentFreq(y, ignoreDC=True, nrange=None):
    N = len(y)
    yf = sm.tsa.stattools.periodogram(y)
    xf = np.linspace(0,1,N) 

    strIndex = 0
    if ignoreDC:
        strIndex = 1
    
    endIndex = N
    if nrange != None:
        endIndex = nrange + strIndex
    
    ibest = np.argmax(yf[strIndex:endIndex]) + strIndex
    
    return xf[ibest]
```
