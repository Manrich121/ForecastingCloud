```python
%matplotlib inline
import fileutils
import numpy as np
from matplotlib import pyplot as plt
import holtwinters as hw
from scipy import fft, arange, sqrt
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import statsmodels as sm
```

```python
data = pd.read_csv("D:/data/cpuRate/cpu_660404.csv")
'''
Input window = 250 hours = 250*12 = 3000 
look ahead window 60 samples =  5 hours = 720min/5 = 60
'''
input_window = 3000
predic_window = 60

y_true = data.Cpu[input_window:input_window+predic_window]
y = data.Cpu[:input_window].tolist()

# y_pred, alpha, beta, gamma, rmse = hw.additive(y, fc=predic_window)
params, rmse = hw.fit(y, type='linear')
alpha, beta, gamma, m = params

print "Model rmse:", rmse
print "alpha:", alpha, "beta", beta, "gamma:", gamma

y_pred = hw.predict(y, fc=predic_window, modelparams=params, type='linear')

print "Forecast RMSE:", mean_squared_error(y_true, y_pred)
print "R sqaured:", r2_score(y_true, y_pred)
```

```python
plt.plot(data.Cpu);
plt.xlim(0, 8500);
plt.figure()

plt.plot(y_true, label='Real');
plt.plot(y_pred, label='Forecast');
plt.legend();
```
