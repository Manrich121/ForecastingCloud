```python
%matplotlib inline
import fileutils
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
from pandas.tools.plotting import lag_plot
import pandas
import tsutils
```

```python
data1 = np.genfromtxt("d:/data/cpuRate/cpu_660404.csv", delimiter=',', usecols=(0,1))

fig = plt.figure(figsize=(12,8))
plt.plot(data1[:,1]);
plt.xlim(0, 8500);
```

### Explore AGILE's method

```python
threshold =  np.percentile(data1[:,1],70)
plt.boxplot(data1[:,1]);
```

```python
overload_samples = data1[data1[:,1]>threshold]
print overload_samples.shape
```

```python
plt.figure(figsize=(12,8));
plt.scatter(overload_samples[:,0],overload_samples[:,1]);
```

```python
overload_dur = 5 #samples
equalFlag = False
condition = data1[:,1]>threshold

for start, stop in tsutilscontiguous_regions(condition):
    if stop - start >= overload_dur:
        segment = data1[:,1][start:stop]
        print start, stop
        print segment
```

```python
plt.figure(figsize=(12,8));
series = pandas.Series(data1[:,1])
lag_plot(series);
```

```python
data2[:,1] = np.nan_to_num(data2[:,1])
fig = plt.figure(figsize=(12,8))

# Plot Autocorrelation function
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data2[:,1], lags=40, ax=ax1)

# Plot Partial Autocorrelation function 
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data2[:,1], lags=40, ax=ax2)
```
