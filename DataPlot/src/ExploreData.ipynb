```python
%matplotlib inline
import fileutils
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
from pandas.tools.plotting import lag_plot
import pandas
import tsutils
```

```python
data1 = np.genfromtxt("d:/data/cpuRate/cpu_660404.csv", delimiter=',', usecols=(0,1))

fig = plt.figure(figsize=(12,8))
plt.plot(data1[:,1]);
plt.xlim(0, 8500);
```

### Explore AGILE's method

```python
overload_dur = 5 #samples
threshold =  np.percentile(data1[:,1],70)
data1_overload = []

for start, stop in tsutils.contiguous_regions(data1[:,1]>threshold):
    if stop - start >= overload_dur:
        segment = data1[:,1][start:stop]
        data1_overload.append([start,stop])
        print start, stop
        print segment
data1_overload = np.array(data1_overload)
```

```python
plt.figure(figsize=(12,8));
series = pandas.Series(data1[:,1])
lag_plot(series);
```
