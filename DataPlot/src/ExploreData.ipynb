### Wavelet analysis

```python
training = data[100:3100]
levels = 5
fc = 30
order = [1, 1, 2, 4, 8, 16]
types = ['db1', 'db2', 'bior1.5', 'sym4']
plt.plot(training)
```

```python
bestType = 'db1'
bestDist = np.inf

for t in types:
    approx_levels = pywt.wavedec(training, wavelet=t, level=levels)
    idx = np.int_(np.linspace(0, training.shape[0]-1, num=len(approx_levels[0])))
    samples = training[idx]
    dist = np.linalg.norm(approx_levels[0]-samples)
    if dist < bestDist:
        bestDist = dist
        bestType = t
print bestType
```

```python
approx_levels = pywt.wavedec(training, wavelet=bestType, level=levels)
for s in approx_levels:
    print s.shape
```

```python
for i in xrange(len(order)):
    model = AR_model(approx_levels[i], order=order[i])
    model.fit()
    approx_levels[i] = np.append(approx_levels[i], model.predict(order[i])[:,0])
```

```python
for s in approx_levels:
    print s.shape
```

```python
pred = pywt.waverec(approx_levels, wavelet=bestType)
plt.plot(pred)
plt.plot(training)
plt.figure()
plt.plot(pred[-32:])
```

### Signature pattern

```python
training = data[30:3030]
Z = tsutils.findDominentSeason(training)
print Z
plt.plot(training);
freq, pwr = scipy.signal.periodogram(training)
plt.figure();
plt.stem(pwr[:20]);
print 1/freq[np.argmax(pwr)]
```

```python
window = 21
ma = np.convolve(training, np.ones(window)/window, 'same')
smooth = scipy.signal.medfilt(training, window)
plt.figure(figsize=(12,8))
plt.plot(training)
plt.plot(smooth, linewidth=1.5)
plt.plot(ma, linewidth=1.5)
print tsutils.findDominentSeason(smooth)
```

```python
Q = np.int(np.ceil(1.0*training.shape[0]/Z))
print Q
idx = range(Z,Q*Z,Z)
P = np.array(np.split(smooth, idx))
Pr = np.array(np.split(training, idx))
warp = P[-1].shape != P[-2].shape
plt.figure(figsize=(20,8))
plt.subplot(1,Q,1)

PC = []
MR = []
for i in range(Q-1-warp):
    plt.subplot(1,Q,i+1)
    plt.plot(P[i])
    PC.append(scipy.stats.pearsonr(P[i], P[i+1])[0])
    MR.append(np.mean(P[i])/np.mean(P[i+1]))
    plt.subplot(1,Q,Q)
plt.plot(P[-1])

# Assume have found Signature
sig = np.average(Pr[:-1],axis=0)
```

```python
PC = np.array(PC)
MR = np.array(MR)
print PC
print MR
print 1.0*sum(PC>0.75)/len(PC)
print 1.0*sum(np.logical_and(MR>=0.9, MR<=1.1))/len(MR)
```

```python
def absDist(x, y):
    return abs(x-y)
```
