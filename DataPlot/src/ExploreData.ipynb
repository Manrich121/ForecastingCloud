```python
%matplotlib inline
import fileutils
import numpy as np
import scipy
from matplotlib import pyplot as plt
import statsmodels
import statsmodels.api as sm
import pandas 
import tsutils
import Markov_model
import mlpy

print "numpy:", np.__version__
print "scipy:", scipy.__version__
print "statsmodels:", statsmodels.__version__
print "pandas:", pandas.__version__


data = np.genfromtxt("d:/data/cpuRate/cpu_1442486585.csv",skip_header=1, delimiter=',',usecols=(1))
# data = np.genfromtxt("d:/data/cpuRate/cpu_1095481.csv",skip_header=1, delimiter=',',usecols=(1))

plt.figure(figsize=(12,8))
plt.plot(data)
```

### Signature pattern

```python
training = data[30:3030]
Z = tsutils.findDominentSeason(training)
print Z
plt.plot(training);
freq, pwr = scipy.signal.periodogram(training)
plt.figure();
plt.stem(pwr[:20]);
print 1/freq[np.argmax(pwr)]
```

```python
window = 21
ma = np.convolve(training, np.ones(window)/window, 'same')
smooth = scipy.signal.medfilt(training, window)
plt.figure(figsize=(12,8))
plt.plot(training)
plt.plot(smooth, linewidth=1.5)
plt.plot(ma, linewidth=1.5)
print tsutils.findDominentSeason(smooth)
```

```python
Q = np.int(np.ceil(1.0*training.shape[0]/Z))
print Q
idx = range(Z,Q*Z,Z)
P = np.array(np.split(ma, idx))
Pr = np.array(np.split(training, idx))
print P[-1].shape, P[-2].shape, P[-1].shape == P[-2].shape
plt.figure(figsize=(20,8))
plt.subplot(1,Q,1)
for i in range(Q-1):
    plt.subplot(1,Q,i+1)
    plt.plot(P[i])
    print "PC=", scipy.stats.pearsonr(P[i], P[i+1])[0], "MeanRatio=", np.mean(P[i])/np.mean(P[i+1])
    plt.subplot(1,Q,Q)
plt.plot(P[-1])

# Assume have found Signature
sig = np.average(Pr[:-1],axis=0)
print sig.shape
# plt.plot(sig)
```

#### DTW

```python
x, y = P[-1], sig
print x.shape, y.shape
plt.plot(x)
plt.plot(y)
plt.figure()

dist, cost, path = mlpy.dtw_std(x, y, dist_only=False)
index = x.shape[0] + np.int(np.ceil(dist))
print "Index on Signature =",index
plt.plot(x[path[0]])
plt.plot(y[path[1]])

plt.figure()
plt.plot(path[0], path[1])
```
