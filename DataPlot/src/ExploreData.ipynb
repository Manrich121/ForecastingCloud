```python
%matplotlib inline
import fileutils
import numpy as np
import scipy
from matplotlib import pyplot as plt
import statsmodels
import statsmodels.api as sm
from pandas.tools.plotting import lag_plot
import pandas 
import tsutils

print "numpy:", np.__version__
print "scipy:", scipy.__version__
print "statsmodels:", statsmodels.__version__
print "pandas:", pandas.__version__
```

```python
data = pandas.read_csv("d:/data/cpuRate/cpu_907812.csv",usecols=[1])
data = data.fillna(0)
data.plot(figsize=(12,8))
```

```python
f = tsutils.findDominentSeason(data.values.squeeze())
print f
res = sm.tsa.seasonal_decompose(data.values.squeeze(), freq=1500);
print res.plot()
```

### ARIMA model

```python
fig = plt.figure(figsize=(12,8))

# Plot Autocorrelation function
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data, lags=50, ax=ax1)

# Plot Partial Autocorrelation function 
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data, lags=50, ax=ax2)
```

```python
ddata = data.diff()
ddata = ddata.fillna(0)
fig = plt.figure(figsize=(12,8))

# Plot Autocorrelation function
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ddata.values[:50], ax=ax1)

# Plot Partial Autocorrelation function 
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ddata.values[:50], ax=ax2)
```

```python
model = statsmodels.tsa.ar_model.AR(data.values.squeeze()[:3000]).fit()

print model.params.shape
plt.plot(range(2900,3060), data.values.squeeze()[2900:3060])
plt.plot(range(3000, 3060), model.predict(3000,3059))
```

```python
y = data.values.squeeze()[:3000]
model = statsmodels.tsa.arima_model.ARIMA(y, order=(10,0,0)).fit(trend='nc')
print model.summary()
plt.plot(model.fittedvalues)
```

```python
%timeit statsmodels.tsa.arima_model.ARIMA(y, order=(10,0,0)).fit()
```
