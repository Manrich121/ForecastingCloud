```python
%matplotlib inline
import numpy as np
import scipy
from matplotlib import pyplot as plt
import tsutils
import Markov_model
```

```python
data = np.nan_to_num(np.genfromtxt("d:/data/cpuRate/cpu_907812.csv", delimiter=',', skip_header=1,usecols=[1]))
data = data[:3000]
M = 40
plt.plot(data)
```

```python
model = Markov_model.Markov_model(data, M=M, order=2)
model.fit()
model.predict(30)
```

```python
bins = np.array(np.linspace(0,max(data),M+1))
states = np.digitize(data,bins, right=True)
# digiData =  bins[idx]

plt.plot(states)
```

```python
trans = np.array([np.zeros((M,M))]*M)
for i in range(1,len(states)-1):
    x, y, z = states[i-1], states[i], states[i+1]
    trans[x-1,y-1,z-1] += 1

trans = np.divide(trans, np.array([np.tile(np.sum(trans[idx], axis=1), (M,1)).T for idx in range(M)]))
trans = np.nan_to_num(trans)
print trans
```

```python
pred = np.zeros((30,1))
prev, cur = states[-2], states[-1]
for p in range(30):
#     print prev, cur
    pred[p,0] = np.argmax(trans[prev-1, cur-1]) +1
    prev = cur
    cur = pred[p,0]

print bins[np.int_(pred.squeeze())]
plt.plot(bins[np.int_(pred.squeeze())])
```
