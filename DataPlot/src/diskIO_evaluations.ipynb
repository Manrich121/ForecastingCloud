```python
%matplotlib inline
import fileutils
import tsutils
import numpy as np
import matplotlib.pyplot as plt

curfile = 'diskIO_907812.csv'
resource = 'diskio'

truevals = np.genfromtxt('d:/data/diskio/'+curfile, delimiter=',',skip_header=1)[:3840,1]
threshold =  np.percentile(truevals, 70)
```

### Holtwinters

```python
forecasts = np.genfromtxt("d:/data/"+resource+"_hw_forecasts/"+curfile,delimiter=',',usecols=range(0,30)).ravel()
plt.figure(figsize=(12,8))
plt.plot(truevals)
plt.plot(range(3000,3000+len(forecasts)),forecasts)
plt.hlines(threshold,0,len(truevals))
```
