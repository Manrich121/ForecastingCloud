```python
def reject_outliers(data, m = 2.):
    data = np.nan_to_num(data)
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

```python
def calcT(x1, s1, x2, s2, n):
    sp = np.sqrt(((n-1)*s1*s1 + (n-1)*s2*s2)/(2.0*n-2))
    return (x1 - x2)/(sp*np.sqrt(2.0/n))

def calcPvalue(data1, data2, N=500):
    x1, s1 = np.mean(data1), np.std(data1)
    x2, s2 = np.mean(data2), np.std(data2)
    print 'pval=', scipy.stats.t.sf(np.abs(calcT(x1, s1, x2, s2, N)), N-1)
```

```python
def calcGain(x1, x2):
    mu1 = np.mean(x1)
    mu2 = np.mean(x2)
    g = 1 - mu2/mu1
    return g
```

```python
plt.figure(figsize=(12,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pos = [0,0.5, 1.5,2, 3,3.5, 4.5,5, 6,6.5, 7.5,8, 9,9.5, 10.5,11]
pltdic = plt.boxplot([reject_outliers(results[i][:,0]) for i in range(len(labs))], positions=pos, widths=0.3, whis=[10,90], showfliers=False, labels=labs);
plt.grid()
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.title("RMSE: Comparing forecasting window lengths on Memory usage data",fontsize=15)
plt.subplots_adjust(bottom=0.2)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/mem_15_rmse.pdf", type='pdf')
gains_rmse = []
for i in [0,2,4,6,8,10,12,14]:
    gains_rmse.append(calcGain(reject_outliers(results[i][:,0]),reject_outliers(results[i+1][:,0])))
print 'mean_gain=', 100*np.mean(gains_rmse)
```
