```python
%matplotlib inline
import numpy as np
import scipy
import statsmodels
import statsmodels.api as sm
from matplotlib import pyplot as plt
import AR_model

data = np.nan_to_num(np.genfromtxt("d:/data/cpuRate/cpu_1095481.csv", delimiter=',', skip_header=1, usecols=(1)))[:3000]
plt.plot(data)
Z = 30
```

```python
R = sm.tsa.stattools.acf(data,nlags=Z)
R = np.r_[R,R[::-1]][:-1]
plt.stem(R)
```

```python
a = np.zeros((Z,Z))
b = np.zeros((Z,1))
for j in range(1,Z+1):
    b[j-1] = R[j]
    for i in range(1,Z+1):
        a[j-1,i-1] = R[i-j]
        
params = np.linalg.solve(a, b)
print params.shape
```

```python
window = Z
mu = np.mean(data)
forecasts = np.zeros((window,1))
x = np.zeros((1,window))
for f in range(window):
    for t in range(window-f):
        x[0,t] = data[-(t+1)]
    if f>0:
        x[0,-f] = forecasts[f,0]
    forecasts[f,0] = np.dot(x-mu, params) + mu
```
