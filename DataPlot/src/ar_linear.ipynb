```python
%matplotlib inline
import numpy as np
import scipy
import statsmodels
import statsmodels.api as sm
from matplotlib import pyplot as plt
import AR_model

data = np.nan_to_num(np.genfromtxt("d:/data/cpu/cpu_1095481.csv", delimiter=',', skip_header=1, usecols=(1)))
train = data[2000:4000]
plt.plot(data)
Z = 30
window = 30
```

```python
R = sm.tsa.stattools.acf(train,nlags=Z)
R = np.r_[R,R[::-1]][:-1]
plt.stem(R)
```

```python
a = np.zeros((Z,Z))
b = np.zeros((Z,1))
for j in range(1,Z+1):
    b[j-1] = R[j]
    for i in range(1,Z+1):
        a[j-1,i-1] = R[i-j]
        
params = np.linalg.solve(a, b)
print params.shape
```

```python
mu = np.mean(train[-window:])
forecasts = []
x = np.zeros((1,Z))
for f in range(window):
    if f == 0:
        x[0,:] = train[-Z:]
    else:
        x[0,:] = np.append(x[0,1:], forecasts[-1])
    fc = mu + np.dot(x - mu, params)
    forecasts.append(fc[0][0])
```

```python
plt.plot(data[3000:3000+window])
plt.plot(forecasts,'-o')
```
