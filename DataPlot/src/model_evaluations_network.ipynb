# Comparing prediction/forecasting models on 8 accuracy metrics, for network.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
ma = np.genfromtxt("d:/wikipage data/results/network_ma.csv", delimiter=',')
ma1 = np.genfromtxt("d:/wikipage data/results/network_ma1.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/wikipage data/results/network_hw.csv", delimiter=',')
ar8 = np.genfromtxt("d:/wikipage data/results/network_ar8.csv", delimiter=',')
ar16 = np.genfromtxt("d:/wikipage data/results/network_ar16.csv", delimiter=',')
ar30 = np.genfromtxt("d:/wikipage data/results/network_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/wikipage data/results/network_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/wikipage data/results/network_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/wikipage data/results/network_press.csv", delimiter=',')
agile = np.genfromtxt("d:/wikipage data/results/network_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/wikipage data/results/network_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/wikipage data/results/network_rnn.csv", delimiter=',')

results = np.array([ma1, ma, holtwinters, ar8, ar16, ar30, markov1, markov2, press, agile, fnn, rnn])
labs = ['MA(1)', 'MA(30)', 'HoltWinters', 'AR(8)', 'AR(16)', 'AR(30)', '1st_Markov', '2nd_Markov', 'PRESS', 'Agile', 'FFNN', 'RNN']
```

### RMSE

Lower is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([reject_outliers(results[i][:,0]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("RMSE on Network data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_rmse.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_rmse.png", type='png')
calcPvalue(results[0][:,0],results[2][:,0])
```

### Correct Est Rate (within +- 10 %)

Higher is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([reject_outliers(results[i][:,1]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Correct Estimation Rate', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Correct Estimation Rate on Network data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_correct.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_correct.png", type='png')
calcPvalue(results[2][:,1],results[7][:,1])
```

###     Estiamtion Score

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([reject_outliers(0.3*results[i][:,2] + 0.7*results[i][:,3]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False,labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Estimation score', fontsize=15)
plt.grid()
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Estimation score on Network data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_estscore.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_estscore.png", type='png')
calcPvalue(0.3*results[2][:,2] + 0.7*results[0][:,3], 0.3*results[3][:,2] + 0.7*results[2][:,3])
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic=plt.boxplot([reject_outliers(results[i][:,4]/results[i][:,5]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Positive Likelihood Ratio', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("OLR on Network data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_olr.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_olr.png", type='png')
calcPvalue(results[0][:,4]/results[0][:,5],results[7][:,4]/results[7][:,5])
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic=plt.boxplot([reject_outliers(results[i][:,6]/results[i][:,7]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Positive Likelihood Ratio', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("OSLR on Network data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_oslr.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_oslr.png", type='png')
calcPvalue(results[2][:,6]/results[2][:,7],results[1][:,6]/results[1][:,7])
```
