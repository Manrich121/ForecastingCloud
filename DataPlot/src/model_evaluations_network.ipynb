# Comparing prediction/forecasting models on 8 accuracy metrics, for network.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
ma = np.genfromtxt("d:/wikipage data/results/network_ma.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/wikipage data/results/network_hw.csv", delimiter=',')
autoreg = np.genfromtxt("d:/wikipage data/results/network_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/wikipage data/results/network_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/wikipage data/results/network_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/wikipage data/results/network_press.csv", delimiter=',')
agile = np.genfromtxt("d:/wikipage data/results/network_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/wikipage data/results/network_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/wikipage data/results/network_rnn.csv", delimiter=',')

results = np.array([ma, holtwinters, autoreg, markov1, markov2, press, agile, fnn, rnn])
labs = ['MA', 'HoltWinters', 'AR(30)', '1st_Markov', '2nd_Markov', 'PRESS', 'Agile', 'FFNN', 'RNN']
```

```python
def elp(pltdic, best='min',skiplast=0):
    miny = np.inf
    mindx = 0
    maxy = -np.inf
    maxdx = 0
    for i in range(len(labs)-skiplast):
        val = pltdic['medians'][i].get_ydata()[0]
        if  val < miny:
            miny = val
            mindx = i
        if val > maxy:
            maxy = val
            maxdx = i

    if best == 'min':
        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')
    else:
        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')

    return ellipse1, ellipse2
```

## Stat test

```python
def calcT(x1, s1, x2, s2, n):
    sp = np.sqrt(((n-1)*s1*s1 + (n-1)*s2*s2)/(2.0*n-2))
    return (x1 - x2)/(sp*np.sqrt(2.0/n))

def calcPvalue(data1, data2, N=500):
    x1, s1 = np.mean(data1), np.std(data1)
    x2, s2 = np.mean(data2), np.std(data2)
    print 'pval=', scipy.stats.t.sf(np.abs(calcT(x1, s1, x2, s2, N)), N-1)
```

### RMSE

Lower is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([results[i][:,0] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("RMSE on Network data", fontsize=15)
# plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_rmse.pdf", type='pdf')
plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_rmse.png", type='png')
calcPvalue(results[0][:,0],results[2][:,0])
```

### Correct Est Rate (within +- 10 %)

Higher is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([results[i][:,1] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Correct Estimation Rate', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Correct Estimation Rate on Network data", fontsize=15)
# plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/network_correct.pdf", type='pdf')
plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/network_correct.png", type='png')
calcPvalue(results[2][:,1],results[7][:,1])
```
