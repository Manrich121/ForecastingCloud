```python
%matplotlib inline
import fileutils
import tsutils
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy
from tabulate import tabulate
```

```python
def elp(pltdic, best='min',skiplast=0):
    miny = np.inf
    mindx = 0
    maxy = -np.inf
    maxdx = 0
    for i in range(len(labs)-skiplast):
        val = pltdic['medians'][i].get_ydata()[0]
        if  val < miny:
            miny = val
            mindx = i
        if val > maxy:
            maxy = val
            maxdx = i

    if best == 'min':
        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')
    else:
        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')

    return ellipse1, ellipse2
```

```python
def calcT(x1, s1, x2, s2, n):
    sp = np.sqrt(((n-1)*s1*s1 + (n-1)*s2*s2)/(2.0*n-2))
    return (x1 - x2)/(sp*np.sqrt(2.0/n))

def calcPvalue(data1, data2, N=500):
    x1, s1 = np.mean(data1), np.std(data1)
    x2, s2 = np.mean(data2), np.std(data2)
    print 'pval=', scipy.stats.t.sf(np.abs(calcT(x1, s1, x2, s2, N)), N-1)
```

```python
def reject_outliers(data, m = 2.):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

```python
ma = np.genfromtxt("d:/data/results/cpu_ma.csv", delimiter=',')
ar = np.genfromtxt("d:/data/results/cpu_ar.csv", delimiter=',')
agile = np.genfromtxt("d:/data/results/cpu_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/cpu_fnn.csv", delimiter=',')
avg = np.genfromtxt("d:/data/results/cpu_avg4.csv", delimiter=',')
wa = np.genfromtxt("d:/data/results/cpu_wa.csv", delimiter=',')
combo = np.genfromtxt("d:/data/results/cpu_combo4.csv", delimiter=',')

for i in range(2,6):
    ma = np.append(ma,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ma.csv", delimiter=','),axis=0)
    ar = np.append(ar,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar.csv", delimiter=','),axis=0)
    fnn = np.append(fnn,np.genfromtxt("d:/data/results/cpu"+str(i)+"_fnn.csv", delimiter=','),axis=0)
    avg = np.append(avg,np.genfromtxt("d:/data/results/cpu"+str(i)+"_avg4.csv", delimiter=','),axis=0)
    wa = np.append(wa,np.genfromtxt("d:/data/results/cpu"+str(i)+"_wa.csv", delimiter=','),axis=0)
    combo = np.append(combo,np.genfromtxt("d:/data/results/cpu"+str(i)+"_combo4.csv", delimiter=','),axis=0)
    
results = [ma, ar, agile, fnn, avg, wa, combo]
labs = ['MA', 'AR(30)', 'Agile', 'FFNN', 'Average Model', 'WA Model', 'Combo Model']
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic = plt.boxplot([results[i][:,0] for i in range(len(results))], whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.26)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic, 'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Ensemble: RMSE on CPU usage data",fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_ens_rmse.pdf", type='pdf')
calcPvalue(results[1][:,0], reject_outliers(results[4][:,0]))
calcPvalue(results[1][:,0], reject_outliers(results[5][:,0]))
calcPvalue(results[1][:,0], reject_outliers(results[6][:,0]))
```
