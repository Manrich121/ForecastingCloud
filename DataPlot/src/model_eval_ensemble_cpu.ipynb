```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy
from tabulate import tabulate
```

```python
ma = np.genfromtxt("d:/data/results/cpu_ma.csv", delimiter=',')
ar = np.genfromtxt("d:/data/results/cpu_ar.csv", delimiter=',')
agile = np.genfromtxt("d:/data/results/cpu_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/cpu_fnn.csv", delimiter=',')
avg = np.genfromtxt("d:/data/results/cpu_avg4.csv", delimiter=',')
wa = np.genfromtxt("d:/data/results/cpu_wa.csv", delimiter=',')
combo = np.genfromtxt("d:/data/results/cpu_combo4.csv", delimiter=',')

for i in range(2,6):
    ma = np.append(ma,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ma.csv", delimiter=','),axis=0)
    ar = np.append(ar,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar.csv", delimiter=','),axis=0)
    fnn = np.append(fnn,np.genfromtxt("d:/data/results/cpu"+str(i)+"_fnn.csv", delimiter=','),axis=0)
    avg = np.append(avg,np.genfromtxt("d:/data/results/cpu"+str(i)+"_avg4.csv", delimiter=','),axis=0)
    wa = np.append(wa,np.genfromtxt("d:/data/results/cpu"+str(i)+"_wa.csv", delimiter=','),axis=0)
    combo = np.append(combo,np.genfromtxt("d:/data/results/cpu"+str(i)+"_combo4.csv", delimiter=','),axis=0)
    
results = [ma, ar, agile, fnn, avg, wa, combo]
labs = ['MA(30)', 'AR(30)', 'Agile', 'FFNN', 'Average Model', 'WA Model', 'Combo Model']
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic = plt.boxplot([reject_outliers(results[i][:,0]) for i in range(len(results))], whis=[10,90], showfliers=False, labels=labs ,showmeans=True);
plt.subplots_adjust(bottom=0.26)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic, 'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Ensemble: RMSE on CPU usage data",fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_ens_rmse.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/cpu_ens_rmse.png", type='png')
calcPvalue(results[1][:,0], reject_outliers(results[4][:,0]))
calcPvalue(results[1][:,0], reject_outliers(results[5][:,0]))
calcPvalue(results[1][:,0], reject_outliers(results[6][:,0]))
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic = plt.boxplot([reject_outliers(results[i][:,1]) for i in range(len(results))], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.26)
plt.ylabel('Correct Estimation Rate', fontsize=15)
plt.grid()
e = elp(pltdic, 'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Ensemble: Correct Estimation Rate \n on CPU usage data",fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_ens_correct.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/cpu_ens_correct.png", type='png')
calcPvalue(results[0][:,1], reject_outliers(results[4][:,1]))
calcPvalue(results[0][:,1], reject_outliers(results[5][:,1]))
calcPvalue(results[0][:,1], reject_outliers(results[6][:,1]))
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([reject_outliers(0.3*results[i][:,2] + 0.7*results[i][:,3]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False,labels=labs);
plt.subplots_adjust(bottom=0.26)
plt.ylabel('Estimation Score', fontsize=15)
plt.grid()
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Ensemble: Estimation score \n on CPU usage data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_ens_estscore.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/cpu_ens_estscore.png", type='png')
calcPvalue(0.3*results[0][:,2] + 0.7*results[0][:,3], reject_outliers(0.3*results[4][:,2] + 0.7*results[4][:,3]))
calcPvalue(0.3*results[0][:,2] + 0.7*results[0][:,3], reject_outliers(0.3*results[5][:,2] + 0.7*results[5][:,3]))
calcPvalue(0.3*results[0][:,2] + 0.7*results[0][:,3], reject_outliers(0.3*results[6][:,2] + 0.7*results[6][:,3]))
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic=plt.boxplot([reject_outliers(results[i][:,4]/results[i][:,5]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.26)
plt.ylabel('Positive Likelihood Ratio', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Ensemble: OLR on CPU usage data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_ens_olr.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/cpu_ens_olr.png", type='png')
calcPvalue(results[1][:,4]/results[1][:,5], reject_outliers(results[4][:,4]/results[4][:,5]))
calcPvalue(results[1][:,4]/results[1][:,5], reject_outliers(results[5][:,4]/results[5][:,5]))
calcPvalue(results[1][:,4]/results[1][:,5], reject_outliers(results[6][:,4]/results[6][:,5]))
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic=plt.boxplot([reject_outliers(results[i][:,6]/results[i][:,7]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Positive Likelihood Ratio', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Ensemble: OSLR on CPU usage data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_ens_oslr.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/cpu_ens_oslr.png", type='png')
calcPvalue(reject_outliers(results[3][:-1,6]/results[i][:,7]), reject_outliers(results[4][:,6]/results[4][:,7]))
calcPvalue(reject_outliers(results[3][:-1,6]/results[i][:,7]), reject_outliers(results[5][:,6]/results[5][:,7]))
calcPvalue(reject_outliers(results[3][:-1,6]/results[i][:,7]), reject_outliers(results[6][:,6]/results[6][:,7]))
```

```python
means_rmse = [str(np.round(np.median(results[i][:,0]),4)) for i in range(len(results))]
means_rmse.insert(0, 'RMSE')
means_correct = [str(np.round(np.median(results[i][:,1]),4)) for i in range(len(results))]
means_correct.insert(0, 'Correct Est.')
means_est_score = [str(np.round(np.median(0.5*results[i][:,2] + 0.5*results[i][:,3]),4)) for i in range(len(results))]
means_est_score.insert(0, 'Est. Score')
means_olr = [str(np.round(np.median(results[i][:,4]/results[i][:,5]),4)) for i in range(len(results))]
means_olr.insert(0, 'OLR')
means_oslr = [str(np.round(np.median(results[i][:,6]/results[i][:,7]),4)) for i in range(len(results))]
means_oslr.insert(0, 'OSLR')
```
