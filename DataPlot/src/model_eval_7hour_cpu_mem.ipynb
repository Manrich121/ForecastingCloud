```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy
```

```python
ar8 = np.genfromtxt("d:/7h_data/results/cpu_ar8.csv", delimiter=',')
ar16 = np.genfromtxt("d:/7h_data/results/cpu_ar16.csv", delimiter=',')
ar30 = np.genfromtxt("d:/7h_data/results/cpu_ar30.csv", delimiter=',')
press = np.genfromtxt("d:/7h_data/results/cpu_press.csv", delimiter=',')

labs = ['AR(8)', 'AR(16)', 'AR(30)','PRESS']
results = np.array([ar8, ar16, ar30, press])
```

## PRESS's metric

```python
plt.figure(figsize=(5,5))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([results[i][:,2] for i in range(len(labs)) ], whis=[10,90], showfliers=False, showmeans=True, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Over-estimation rate (>10%)', fontsize=15)
plt.grid()
plt.title("Over-estimation rate on CPU usage data \n(7 hour data)",fontsize=15)
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_press_oer.pdf", type='pdf')
```

```python
plt.figure(figsize=(5,5))
plt.xticks(horizontalalignment='left',rotation=-70,fontsize=15)
pltdic= plt.boxplot([results[i][:,3] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs, showmeans=True);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Under-estimation rate (<10%)', fontsize=15)
plt.grid()
plt.title("Under-estimation rate on CPU usage data \n(7 hour data)",fontsize=15)
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_press_uer.pdf", type='pdf')
```
