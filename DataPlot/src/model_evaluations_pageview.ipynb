# Comparing prediction/forecasting models on 8 accuracy metrics, for pageviews.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
ma1 = np.genfromtxt("d:/wikipage data/results/pageviews_ma1.csv", delimiter=',')
ma = np.genfromtxt("d:/wikipage data/results/pageviews_ma.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/wikipage data/results/pageviews_hw.csv", delimiter=',')
ar8 = np.genfromtxt("d:/wikipage data/results/pageviews_ar8.csv", delimiter=',')
ar16 = np.genfromtxt("d:/wikipage data/results/pageviews_ar16.csv", delimiter=',')
ar30 = np.genfromtxt("d:/wikipage data/results/pageviews_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/wikipage data/results/pageviews_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/wikipage data/results/pageviews_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/wikipage data/results/pageviews_press.csv", delimiter=',')
agile = np.genfromtxt("d:/wikipage data/results/pageviews_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/wikipage data/results/pageviews_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/wikipage data/results/pageviews_rnn.csv", delimiter=',')

results = np.array([ma1, ma, holtwinters, ar8, ar16, ar30, markov1, markov2, press, agile, fnn, rnn])
labs = ['MA(1)', 'MA(30)', 'HoltWinters', 'AR(8)', 'AR(16)','AR(30)', '1st_Markov', '2nd_Markov', 'PRESS', 'Agile', 'FFNN', 'RNN']
```

### RMSE

Lower is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([reject_outliers(results[i][:,0]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("RMSE on Pageview data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/pageviews_rmse.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/page_rmse.png", type='png')
# Stat test
calcPvalue(results[0][:,0], results[2][:,2])
```

### Correct Est rate(within +- 10 %)

Higher is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic= plt.boxplot([reject_outliers(results[i][:,1]) for i in range(len(labs)) ], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Correct Estimation Rate', fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("Correct Estimation Rate on Pageview data", fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/pageviews_correct.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/page_correct.png", type='png')
calcPvalue(results[2][:,1],results[7][:,1])
```
