# Comparing prediction/forecasting models on 8 accuracy metrics, for pageviews.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
holtwinters = np.genfromtxt("d:/wikipage data/results/pageviews_hw.csv", delimiter=',')

results = np.array([holtwinters])
labs = ['HoltWinters']
```

### RMSE

Lower is better

```python
plt.figure(figsize=(12,6))
plt.boxplot([results[i,:,0] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
```

### Correct Prediction (within +- 10 %)

Higher is better

```python
plt.figure(figsize=(12,6))
plt.boxplot([results[i,:,1] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
```

### Over-estimation rate (>10% of actual value)

Lower is better

```python
plt.figure(figsize=(12,6))
plt.boxplot([results[i,:,2] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
```

### Under-estimation rate (<10% of actual value)

Lower is better

```python
plt.figure(figsize=(12,6))
plt.boxplot([results[i,:,3] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
```
