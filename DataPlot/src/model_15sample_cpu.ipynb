### Investigate Offset 30 sample forecasting window

```python
curFile = 'cpu_1335648'
truevals = np.genfromtxt("d:/data/cpu/"+curFile+".csv", delimiter=',',skip_header=1,usecols=(1))
forecasts= np.genfromtxt("d:/data/cpu_ar/"+curFile+".csv",delimiter=',',usecols=range(0,30)).ravel()
# forecasts = np.genfromtxt("D:/15_sample_results/cpu_ma/"+curFile+".csv",delimiter=',',usecols=range(0,15)).ravel()
plt.figure(figsize=(12,8))
plt.plot(truevals)
plt.plot(range(3000,3000+len(forecasts)),forecasts,linewidth=1.5)
```

```python
def findOffset(truevals, forecasts, metric, best='min', OFFSET_RANGE=30):
    bestOffset = 0
    if best == 'min':
        bestErr = np.inf
    else:
        bestErr = -np.inf
       
    for offset in range(OFFSET_RANGE+1):
        newForecast = np.zeros(len(forecasts))
        if offset == 0:
            newForecast = forecasts
        elif offset <= OFFSET_RANGE/2:
            newForecast[:-offset] = forecasts[offset:]
        else:
            newForecast[(offset-OFFSET_RANGE/2):] = forecasts[:-(offset-OFFSET_RANGE/2)]

        if metric == 'rmse':
            err = eval.calc_RMSE(truevals[3000:3000+len(newForecast)], newForecast)
        elif metric == 'correct':
            err = eval.calc_upper_lower_acc(truevals[3000:3000+len(newForecast)], newForecast)[0]
        elif metric =='estscore':
            err = 0.5*eval.calc_upper_lower_acc(truevals[3000:3000+len(newForecast)], newForecast)[1] + 0.5*eval.calc_upper_lower_acc(truevals[3000:3000+len(newForecast)], newForecast)[2]
        elif metric == 'olr':
            err = eval.calc_persample_accuracy(truevals[3000:3000+len(newForecast)], newForecast, np.percentile(truevals, 70))[0]/eval.calc_persample_accuracy(truevals[3000:3000+len(newForecast)], newForecast,np.percentile(truevals, 70))[1]
        elif metric =='oslr':
            err = eval.calc_overload_states_acc(truevals[3000:3000+len(newForecast)], newForecast, np.percentile(truevals, 70))[0]/eval.calc_overload_states_acc(truevals[3000:3000+len(newForecast)], newForecast, np.percentile(truevals, 70))[1]

        if best == 'min':
            if err < bestErr:
                bestErr = err
                if offset <= OFFSET_RANGE/2:
                    bestOffset = -1*offset
                else:
                    bestOffset = offset-OFFSET_RANGE/2
            
        if best == 'max':
            if err > bestErr:
                bestErr = err
                if offset <= OFFSET_RANGE/2:
                    bestOffset = -1*offset
                else:
                    bestOffset = offset-OFFSET_RANGE/2
                
    return bestErr, bestOffset
```
