```python
%matplotlib inline
import fileutils
import numpy as np
import scipy
from matplotlib import pyplot as plt
import statsmodels
import statsmodels.api as sm
from collections import defaultdict
import string
```

```python
files = fileutils.getFilelist("D:\Wikipage data\projectcounts-2014")
input_data = np.genfromtxt(files[0], delimiter=' ', usecols=(0,2,3), dtype='string')
```

```python
# top100 = np.flipud(input_data[np.argsort(np.int_(input_data[:,1]))[-100:],:])
top100 = np.flipud(input_data[np.argsort(np.int_(input_data[:,1])),:])
print top100.shape
top100labels = top100[:,0]
```

```python
topLang = []
all_labels = np.flipud(input_data[np.argsort(np.int_(input_data[:,1])),0])
count = 0
for l in all_labels:
    if count < 100:
        if string.count(l, '.') == 0 :
            topLang.append(l)
            count += 1
    else:
        break
    
```

```python
print topLang
```

```python
views = defaultdict(list)
network = defaultdict(list)
for f in files:
#     print f, "started"
    data = np.genfromtxt(f, delimiter=' ', usecols=(0,2,3), dtype='string')
    for l in topLang:
        if np.any(data[:,0]==l):
            views[l].append(np.double(data[data[:,0]==l,1]))
            network[l].append(np.double(data[data[:,0]==l,2]))
```

```python
print plt.plot(network['en']/np.max(network['en']))
```

```python
for lang in topLang:
    fileutils.writeCSV("D:/Wikipage data/pageviews/view_"+lang, np.atleast_2d(views[lang]))
    fileutils.writeCSV("D:/Wikipage data/network/network_"+lang, np.atleast_2d(network[lang]))
```
