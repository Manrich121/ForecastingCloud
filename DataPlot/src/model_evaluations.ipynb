# Comparing prediction/forecasting models on 8 accuracy metrics.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
norm = np.genfromtxt("d:/data/results/normaldist.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/data/results/holtwinters.csv", delimiter=',')
autoreg = np.genfromtxt("d:/data/results/autoregressive.csv", delimiter=',')
labs = ['NormalDistr','Holtwinters', 'AutoRegression']
```

### RMSE

Lower is better

```python
plt.boxplot((norm[:,0], holtwinters[:,0], autoreg[:,0]), whis=[10,90], showfliers=False, labels=labs);
```

### Correct Prediction (within +- 10 %)

Higher is better
