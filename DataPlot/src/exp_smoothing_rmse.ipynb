```python
%matplotlib inline
import fileutils
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
```

```python
data = np.genfromtxt("d:/data/cpuRate/cpu_exp_smoothing_rmse.csv", delimiter=',')
```

```python
plt.figure(figsize=(30,10))
plt.boxplot(data.T);
plt.title("RMSE Errors of all 100 machines");
```

From the figure we observe that except for machine 69 and 74, the RMSE is relatively simmilar across all machines


Now we take the average RMSE error for each machine and plot it. Again we see machine 69 and 74 having the highest average RMSE

```python
aggregated = np.mean(data,axis=1)
plt.figure(figsize=(30,10));
plt.stem(np.arange(1,101),aggregated);
plt.xticks(np.arange(1,101));
```

We now make a boxplot of the the averages, with the bottom and top of the box representing 25th and 75th percentile and the ends of the whiskers the 10th and 90th percentile.

```python
plt.boxplot(aggregated,whis=[10,90], showfliers=False,labels=["Exp_Smoothing"]);
plt.title("Average Root Mean Squared Error (RMSE) for Exponetial Smoothing");
```
