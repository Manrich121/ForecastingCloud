```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy
```

```python
ar8 = np.genfromtxt("d:/data/results/cpu_ar8.csv", delimiter=',')
ar16 = np.genfromtxt("d:/data/results/cpu_ar16.csv", delimiter=',')
ar30 = np.genfromtxt("d:/data/results/cpu_ar.csv", delimiter=',')
press = np.genfromtxt("d:/data/results/cpu_press.csv", delimiter=',')
agile = np.genfromtxt("d:/data/results/cpu_agile.csv", delimiter=',')

for i in [2,3,4,5]:

    ar8=np.append(ar8,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar8.csv", delimiter=','),axis=0)
    ar16=np.append(ar16,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar16.csv", delimiter=','),axis=0)
    ar30=np.append(ar30,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar.csv", delimiter=','),axis=0)
    
    press=np.append(press,np.genfromtxt("d:/data/results/cpu"+str(i)+"_press.csv", delimiter=','),axis=0)
    agile=np.append(agile,np.genfromtxt("d:/data/results/cpu"+str(i)+"_agile.csv", delimiter=','),axis=0)
    
results = np.array([ar8, ar16, ar30, press, agile])
labs = ['AR(8)', 'AR(16)', 'AR(30)', 'PRESS', 'Agile']
```

# Agile Metrics

## CPU

### True Positive Rate for overloaded observation

```python
plt.figure(figsize=(5,5))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic=plt.boxplot([reject_outliers(results[i][:,4]) for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs,showmeans=True);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('True Positive Rate', fontsize=15)
plt.title("TPR for Overloaded Observations\n on CPU usage data",fontsize=15)
plt.grid()
e = elp(pltdic,'max',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_agile_tp.pdf", type='pdf')
```

### False Positive Rate for Overloaded Observations

```python
plt.figure(figsize=(5,5))
plt.xticks(horizontalalignment='left',rotation=-70, size='large')
pltdic=plt.boxplot([reject_outliers(results[i][:,5]) for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs,showmeans=True);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('False Positive Rate', fontsize=15 )
plt.title("FPR for Overloaded Observations\n on CPU usage data",fontsize=15)
plt.grid()
e = elp(pltdic,'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu_agile_fp.pdf", type='pdf')
```
