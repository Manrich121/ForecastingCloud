```python
%matplotlib inline
# from __future__ import print_function

import numpy as np
import scipy
from matplotlib import pyplot as plt
import evaluation as eval

from pybrain.datasets import SupervisedDataSet
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.structure.modules import SigmoidLayer
```

```python
data = np.genfromtxt("d:/data/cpuRate/cpu_5656207113.csv",skip_header=1, delimiter=',',usecols=(1))
threshold =  np.percentile(data, 70)
plt.figure(figsize=(12,8))
plt.plot(data)
```

```python
TRAIN = 1000
VALID = 100
TEST = 100
INPUT_SIZE = 30
HIDDEN = 10
train, valid, test = data[:TRAIN], data[TRAIN-INPUT_SIZE:TRAIN+VALID], data[TRAIN-INPUT_SIZE+VALID:TRAIN+VALID+TEST]
```

```python
plt.plot(train)
```

```python
trainds = SupervisedDataSet(INPUT_SIZE, 1)
testds = SupervisedDataSet(INPUT_SIZE, 1)
validds = SupervisedDataSet(INPUT_SIZE, 1)

for i in range(INPUT_SIZE,train.shape[0]):
    trainds.appendLinked(train[i-INPUT_SIZE:i],train[i])

for i in range(INPUT_SIZE,test.shape[0]):
    testds.appendLinked(test[i-INPUT_SIZE:i],test[i])

for i in range(INPUT_SIZE,valid.shape[0]):
    validds.appendLinked(valid[i-INPUT_SIZE:i],valid[i])
```

```python
ffnet = buildNetwork(INPUT_SIZE, HIDDEN, 1,  bias=True)
print ffnet
```

```python
trainer = BackpropTrainer(ffnet, trainds, learningrate=0.0001, weightdecay=0.01, shuffle=False, verbose=True)
train_err, valid_err = trainer.trainUntilConvergence(maxEpochs=100, validationData=validds)
```

```python
SPLIT = 10
plt.figure()
plt.plot(train_err[:SPLIT])
plt.plot(valid_err[:SPLIT])
plt.figure()
plt.plot(train_err[SPLIT:])
plt.plot(valid_err[SPLIT:])
```

## Training Predictions

```python
pred = ffnet.activateOnDataset(trainds)
truevals = trainds['target']
eval.calc_RMSE(truevals, pred, print_=True)
eval.calc_upper_lower_acc(truevals, pred, print_=True)

plt.figure(figsize=(12,8))
plt.plot(truevals[-100:])
plt.plot(pred[-100:], '-o')
```

## Validation Predictions

```python
pred = ffnet.activateOnDataset(validds)
truevals = validds['target']
eval.calc_RMSE(truevals, pred, print_=True)
eval.calc_upper_lower_acc(truevals, pred, print_=True)

plt.figure(figsize=(12,8))
plt.plot(truevals)
plt.plot(pred, '-o')
```

## Testing Predictions

```python
pred = ffnet.activateOnDataset(testds)
truevals = testds['target']
eval.calc_RMSE(truevals, pred, print_=True)
eval.calc_upper_lower_acc(truevals, pred, print_=True)

plt.figure(figsize=(12,8))
plt.plot(truevals)
plt.plot(pred, '-o')
```
