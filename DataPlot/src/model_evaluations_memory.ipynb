# Comparing prediction/forecasting models on 8 accuracy metrics, for Memory usage.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy
from tabulate import tabulate

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
ma30 = np.genfromtxt("d:/data/results/memory_ma.csv", delimiter=',')
ma1 = np.genfromtxt("d:/data/results/memory_ma1.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/data/results/memory_holtwinters.csv", delimiter=',')
ar8 = np.genfromtxt("d:/data/results/memory_ar8.csv", delimiter=',')
ar16 = np.genfromtxt("d:/data/results/memory_ar16.csv", delimiter=',')
ar30 = np.genfromtxt("d:/data/results/memory_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/data/results/memory_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/data/results/memory_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/data/results/memory_press.csv", delimiter=',')
agile = np.genfromtxt("d:/data/results/memory_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/memory_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/data/results/memory_rnn.csv", delimiter=',')
avg = np.genfromtxt("d:/data/results/memory_avg.csv", delimiter=',')
combo = np.genfromtxt("d:/data/results/memory_combo.csv", delimiter=',')
labs = ['MA(1)', 'MA(30)', 'HoltWinters', 'AR(8)','AR(16)','AR(30)', '1st_Markov', '2nd_Markov', 'PRESS', 'Agile', 'FFNN', 'RNN']#, 'AvgModel', 'Combined']
```

```python
for i in range(2,6):
    ma1 = np.append(ma1,np.genfromtxt("d:/data/results/memory"+str(i)+"_ma1.csv", delimiter=','),axis=0)
    ma30 = np.append(ma30,np.genfromtxt("d:/data/results/memory"+str(i)+"_ma.csv", delimiter=','),axis=0)
    holtwinters=np.append(holtwinters,np.genfromtxt("d:/data/results/memory"+str(i)+"_hw.csv", delimiter=','),axis=0)
    ar8=np.append(ar8,np.genfromtxt("d:/data/results/memory"+str(i)+"_ar8.csv", delimiter=','),axis=0)
    ar16=np.append(ar16,np.genfromtxt("d:/data/results/memory"+str(i)+"_ar16.csv", delimiter=','),axis=0)
    ar30=np.append(ar30,np.genfromtxt("d:/data/results/memory"+str(i)+"_ar.csv", delimiter=','),axis=0)
    markov1=np.append(markov1,np.genfromtxt("d:/data/results/memory"+str(i)+"_markov1.csv", delimiter=','),axis=0)
    markov2=np.append(markov2,np.genfromtxt("d:/data/results/memory"+str(i)+"_markov2.csv", delimiter=','),axis=0)
    press=np.append(press,np.genfromtxt("d:/data/results/memory"+str(i)+"_press.csv", delimiter=','),axis=0)
    agile=np.append(agile,np.genfromtxt("d:/data/results/memory"+str(i)+"_agile.csv", delimiter=','),axis=0)
    fnn=np.append(fnn,np.genfromtxt("d:/data/results/memory"+str(i)+"_fnn.csv", delimiter=','),axis=0)
    rnn=np.append(rnn,np.genfromtxt("d:/data/results/memory"+str(i)+"_rnn.csv", delimiter=','),axis=0)
    
results = np.array([ma1, ma30, holtwinters, ar8, ar16, ar30, markov1, markov2, press, agile, fnn, rnn])
```

### RMSE

Lower is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic = plt.boxplot([reject_outliers(results[i][:,0]) for i in range(len(labs))], showmeans=True, whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Root Mean Squared Error',fontsize=15)
plt.title("RMSE on Memory usage data",fontsize=15)
plt.grid()
e = elp(pltdic, 'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/mem5x_rmse.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/mem_rmse.png", type='png')
# Stat test
calcPvalue(results[0][:,0],results[2][:,0])
```

## Ensemble model

```python
ma = np.genfromtxt("d:/data/results/memory_ma.csv", delimiter=',')
ar = np.genfromtxt("d:/data/results/memory_ar.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/memory_fnn.csv", delimiter=',')
avg = np.genfromtxt("d:/data/results/memory_avg.csv", delimiter=',')
combo = np.genfromtxt("d:/data/results/memory_combo.csv", delimiter=',')
for i in range(2,6):
    ma = np.append(ma,np.genfromtxt("d:/data/results/memory"+str(i)+"_ma.csv", delimiter=','),axis=0)
    ar = np.append(ar,np.genfromtxt("d:/data/results/memory"+str(i)+"_ar.csv", delimiter=','),axis=0)
    fnn = np.append(fnn,np.genfromtxt("d:/data/results/memory"+str(i)+"_fnn.csv", delimiter=','),axis=0)
    avg = np.append(avg,np.genfromtxt("d:/data/results/memory"+str(i)+"_avg.csv", delimiter=','),axis=0)
    combo = np.append(combo,np.genfromtxt("d:/data/results/memory"+str(i)+"_combo.csv", delimiter=','),axis=0)
    
results = [ma, ar, fnn, avg, combo]
labs = ['MA', 'AR(30)', 'FFNN', 'AverageModel', 'ComboModel']
```

```python
means_rmse = [str(np.round(np.median(results[i][:,0]),4)) for i in range(len(results))]
means_rmse.insert(0, 'RMSE')
means_correct = [str(np.round(np.median(results[i][:,1]),4)) for i in range(len(results))]
means_correct.insert(0, 'Correct Est.')
means_est_score = [str(np.round(np.median(0.5*results[i][:,2] + 0.5*results[i][:,3]),4)) for i in range(len(results))]
means_est_score.insert(0, 'Est. Score')
means_olr = [str(np.round(np.median(results[i][:,4]/results[i][:,5]),4)) for i in range(len(results))]
means_olr.insert(0, 'OLR')
means_oslr = [str(np.round(np.median(results[i][:,6]/results[i][:,7]),4)) for i in range(len(results))]
means_oslr.insert(0, 'OSLR')

labs.insert(0, 'Metric')
```
