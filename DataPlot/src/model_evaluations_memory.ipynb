# Comparing prediction/forecasting models on 8 accuracy metrics, for Memory usage.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
holtwinters = np.genfromtxt("d:/data/results/memory_holtwinters.csv", delimiter=',')
autoreg = np.genfromtxt("d:/data/results/memory_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/data/results/memory_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/data/results/memory_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/data/results/memory_press.csv", delimiter=',')
agile = np.genfromtxt("d:/data/results/memory_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/memory_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/data/results/memory_rnn.csv", delimiter=',')
avg = np.genfromtxt("d:/data/results/memory_avg.csv", delimiter=',')
combo = np.genfromtxt("d:/data/results/memory_combo.csv", delimiter=',')
labs = ['HoltWinters', 'AR(30)', '1st_Markov', '2nd_Markov', 'PRESS', 'Agile', 'FFNN', 'RNN']#, 'AvgModel', 'Combined']
```

```python
for i in range(2,6):
    holtwinters=np.append(holtwinters,np.genfromtxt("d:/data/results/memory"+str(i)+"_hw.csv", delimiter=','),axis=0)
    autoreg=np.append(autoreg,np.genfromtxt("d:/data/results/memory"+str(i)+"_ar.csv", delimiter=','),axis=0)
    markov1=np.append(markov1,np.genfromtxt("d:/data/results/memory"+str(i)+"_markov1.csv", delimiter=','),axis=0)
    markov2=np.append(markov2,np.genfromtxt("d:/data/results/memory"+str(i)+"_markov1.csv", delimiter=','),axis=0)
    press=np.append(press,np.genfromtxt("d:/data/results/memory"+str(i)+"_press.csv", delimiter=','),axis=0)
    agile=np.append(agile,np.genfromtxt("d:/data/results/memory"+str(i)+"_agile.csv", delimiter=','),axis=0)
    fnn=np.append(fnn,np.genfromtxt("d:/data/results/memory"+str(i)+"_fnn.csv", delimiter=','),axis=0)
    rnn=np.append(rnn,np.genfromtxt("d:/data/results/memory"+str(i)+"_rnn.csv", delimiter=','),axis=0)
    
results = np.array([holtwinters, autoreg, markov1, markov2, press, agile, fnn, rnn])#, avg, combo])
```

```python
def elp(pltdic, best='min',skiplast=0):
    miny = np.inf
    mindx = 0
    maxy = -np.inf
    maxdx = 0
    for i in range(len(labs)-skiplast):
        val = pltdic['medians'][i].get_ydata()[0]
        if  val < miny:
            miny = val
            mindx = i
        if val > maxy:
            maxy = val
            maxdx = i

    if best == 'min':
        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')
    else:
        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')

    return ellipse1, ellipse2
```

### RMSE

Lower is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=16)
pltdic = plt.boxplot([results[i][:,0] for i in range(len(labs))], whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.22)
plt.ylabel('Root Mean Squared Error',size='large')
plt.grid()
e = elp(pltdic, 'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/data/results_figs/mem5x_rmse.pdf", type='pdf')
```
