# Comparing prediction/forecasting models on 8 accuracy metrics, for CPU usage.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt

'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
# norm = np.genfromtxt("d:/data/results/cpu_normaldist.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/data/results/cpu_holtwinters.csv", delimiter=',')
autoreg = np.genfromtxt("d:/data/results/cpu_ar.csv", delimiter=',')
ar2 = np.genfromtxt("d:/data/results/cpu2_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/data/results/cpu_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/data/results/cpu_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/data/results/cpu_press.csv", delimiter=',')
press2 = np.genfromtxt("d:/data/results/cpu2_press.csv", delimiter=',')
alige = np.genfromtxt("d:/data/results/cpu_agile.csv", delimiter=',')
agile2 = np.genfromtxt("d:/data/results/cpu2_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/cpu_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/data/results/cpu_rnn.csv", delimiter=',')
avg = np.genfromtxt("d:/data/results/cpu_avg.csv", delimiter=',')
combo = np.genfromtxt("d:/data/results/cpu_combo.csv", delimiter=',')
results = np.array([holtwinters, autoreg, ar2, markov1, markov2, press, press2, alige, agile2, fnn, rnn, avg, combo])
labs = ['HoltWinters', 'AR(30)', 'AR2', '1st_Markov', '2nd_Markov', 'PRESS', 'Press2', 'Agile', 'Agile2', 'FFNN', 'RNN', 'AvgModel', 'Combined']
```

### RMSE

Lower is better

```python
plt.figure(figsize=(12,6))
plt.boxplot([results[i,:,0] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
plt.show()
```

```python
plt.xticks(horizontalalignment='left',rotation=-70, size='large')
plt.boxplot([results[i,:,0] for i in range(len(labs)) ], whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.22)
plt.ylabel('Root Mean Squared Error',size='large')
plt.show()
```
