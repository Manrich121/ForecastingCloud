### RMSE

Lower is better

```python
def elp(pltdic, best='min',skiplast=0):
    miny = np.inf
    mindx = 0
    maxy = -np.inf
    maxdx = 0
    for i in range(len(labs)-skiplast):
        val = pltdic['medians'][i].get_ydata()[0]
        if  val < miny:
            miny = val
            mindx = i
        if val > maxy:
            maxy = val
            maxdx = i

    if best == 'min':
        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')
    else:
        y1,y2 = pltdic['caps'][maxdx*2].get_ydata()[0], pltdic['caps'][maxdx*2+1].get_ydata()[0]
        ellipse1 = mpatches.Ellipse((maxdx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='green',linewidth=1.5)

        y1,y2 = pltdic['caps'][mindx*2].get_ydata()[0], pltdic['caps'][mindx*2+1].get_ydata()[0]
        ellipse2 = mpatches.Ellipse((mindx+1.0, (y1+y2)/2), 0.7, (y2-y1)*1.5, fill=False, ec='red',linewidth=1.5, linestyle='dashed')

    return ellipse1, ellipse2
```

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic = plt.boxplot([results[i][:,0] for i in range(len(labs))], whis=[10,90], showfliers=False, labels=labs);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic, 'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.savefig("D:/data/results_figs/cpu5x_rmse.pdf", type='pdf')
plt.show()
```

```python
ar = np.genfromtxt("d:/data/results/cpu_ar.csv", delimiter=',')
ar = np.append(ar,np.genfromtxt("d:/data/results/cpu2_ar.csv", delimiter=','),axis=0)
ar = np.append(ar,np.genfromtxt("d:/data/results/cpu3_ar.csv", delimiter=','),axis=0)
ar = np.append(ar,np.genfromtxt("d:/data/results/cpu4_ar.csv", delimiter=','),axis=0)
ar = np.append(ar,np.genfromtxt("d:/data/results/cpu5_ar.csv", delimiter=','),axis=0)
ar= np.nan_to_num(ar)

agile = np.genfromtxt("d:/data/results/cpu_agile.csv", delimiter=',')
agile= np.append(agile,np.genfromtxt("d:/data/results/cpu2_agile.csv", delimiter=','),axis=0)
agile= np.append(agile,np.genfromtxt("d:/data/results/cpu3_agile.csv", delimiter=','),axis=0)
agile= np.append(agile,np.genfromtxt("d:/data/results/cpu4_agile.csv", delimiter=','),axis=0)
agile= np.append(agile,np.genfromtxt("d:/data/results/cpu5_agile.csv", delimiter=','),axis=0)
agile= np.nan_to_num(agile)

press = np.genfromtxt("d:/data/results/cpu_press.csv", delimiter=',')
press= np.append(press,np.genfromtxt("d:/data/results/cpu2_press.csv", delimiter=','),axis=0)
press= np.append(press,np.genfromtxt("d:/data/results/cpu3_press.csv", delimiter=','),axis=0)
press= np.append(press,np.genfromtxt("d:/data/results/cpu4_press.csv", delimiter=','),axis=0)
press= np.append(press,np.genfromtxt("d:/data/results/cpu5_press.csv", delimiter=','),axis=0)
press = np.nan_to_num(press)

print ar.shape, agile.shape, press.shape
results = np.array([ar, press, agile])
labs = ['AR(30)', 'PRESS', 'Agile']
```

```python
def calcT(x1, s1, x2, s2, n):
    sp = np.sqrt(((n-1)*s1*s1 + (n-1)*s2*s2)/(2.0*n-2))
    return (x1 - x2)/(sp*np.sqrt(2.0/n))
```

```python
metric = ['RMSE', 'Correct', 'Overest', 'Underest', 'TPR', 'FPR', 'TPR_state', 'FPR_state']
N = 500
for i in range(8):
    x1, s1 = np.mean(ar[:,i]), np.std(ar[:,i])
    x2, s2 = np.mean(press[:,i]), np.std(press[:,i])
    print metric[i],'pval=', scipy.stats.t.sf(np.abs(calcT(x1, s1, x2, s2, N)), N-1)
```

```python
for i in range(8):
    x1, s1 = np.mean(ar[:,i]), np.std(ar[:,i])
    x2, s2 = np.mean(agile[:,i]), np.std(agile[:,i])
    print metric[i],'pval=', scipy.stats.t.sf(np.abs(calcT(x1, s1, x2, s2, N)), N-1)
```
