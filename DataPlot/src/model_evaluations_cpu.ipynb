# Comparing prediction/forecasting models on 8 accuracy metrics, for CPU usage.

#### Boxplot: Center line represent mean, top and bottom of box represents the 25the and 75th percentile and the bottom and top whisker represent the 10th and 90th percentile.

```python
%matplotlib inline
import fileutils
import tsutils
from tsutils import elp, reject_outliers, calcPvalue
import evaluation as eval
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy
from tabulate import tabulate
```

```python
'''
Results are saved to file in the following format
[RMSE, correct_rate, overest_rate, underest_rate, TPR_persample, FPR_persample, TPR_overloadstate, FPR_overloadstat]
'''
# norm = np.genfromtxt("d:/data/results/cpu_normaldist.csv", delimiter=',')
ma = np.genfromtxt("d:/data/results/cpu_ma.csv", delimiter=',')
ma1 = np.genfromtxt("d:/data/results/cpu_ma1.csv", delimiter=',')
holtwinters = np.genfromtxt("d:/data/results/cpu_hw.csv", delimiter=',')
ar8 = np.genfromtxt("d:/data/results/cpu_ar8.csv", delimiter=',')
ar16 = np.genfromtxt("d:/data/results/cpu_ar16.csv", delimiter=',')
ar30 = np.genfromtxt("d:/data/results/cpu_ar.csv", delimiter=',')
markov1 = np.genfromtxt("d:/data/results/cpu_markov1.csv", delimiter=',')
markov2 = np.genfromtxt("d:/data/results/cpu_markov2.csv", delimiter=',')
press = np.genfromtxt("d:/data/results/cpu_press.csv", delimiter=',')
agile = np.genfromtxt("d:/data/results/cpu_agile.csv", delimiter=',')
fnn = np.genfromtxt("d:/data/results/cpu_fnn.csv", delimiter=',')
rnn = np.genfromtxt("d:/data/results/cpu_rnn.csv", delimiter=',')

labs = ['MA(1)', 'MA(30)', 'HoltWinters', 'AR(8)', 'AR(16)', 'AR(30)', '1st_Markov', '2nd_Markov', 'PRESS', 'Agile', 'FFNN', 'RNN']#, 'AvgModel', 'Combined']
```

```python
for i in [2,3,4,5]:
    ma1 = np.append(ma1,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ma1.csv", delimiter=','),axis=0)
    ma = np.append(ma,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ma.csv", delimiter=','),axis=0)
    holtwinters=np.append(holtwinters,np.genfromtxt("d:/data/results/cpu"+str(i)+"_hw.csv", delimiter=','),axis=0)
    ar30=np.append(ar30,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar.csv", delimiter=','),axis=0)
    ar8=np.append(ar8,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar8.csv", delimiter=','),axis=0)
    ar16=np.append(ar16,np.genfromtxt("d:/data/results/cpu"+str(i)+"_ar16.csv", delimiter=','),axis=0)
    markov1=np.append(markov1,np.genfromtxt("d:/data/results/cpu"+str(i)+"_markov1.csv", delimiter=','),axis=0)
    markov2=np.append(markov2,np.genfromtxt("d:/data/results/cpu"+str(i)+"_markov2.csv", delimiter=','),axis=0)
    press=np.append(press,np.genfromtxt("d:/data/results/cpu"+str(i)+"_press.csv", delimiter=','),axis=0)
    agile=np.append(agile,np.genfromtxt("d:/data/results/cpu"+str(i)+"_agile.csv", delimiter=','),axis=0)
    fnn=np.append(fnn,np.genfromtxt("d:/data/results/cpu"+str(i)+"_fnn.csv", delimiter=','),axis=0)
    rnn=np.append(rnn,np.genfromtxt("d:/data/results/cpu"+str(i)+"_rnn.csv", delimiter=','),axis=0)
    
results = np.array([ma1, ma, holtwinters, ar8, ar16, ar30, markov1, markov2, press, agile, fnn, rnn])
```

### RMSE

Lower is better

```python
plt.figure(figsize=(6,6))
plt.xticks(horizontalalignment='left',rotation=-70, fontsize=15)
pltdic = plt.boxplot([reject_outliers(results[i][:,0]) for i in range(len(labs))], whis=[10,90], showfliers=False, labels=labs, showmeans=True);
plt.subplots_adjust(bottom=0.24)
plt.ylabel('Root Mean Squared Error', fontsize=15)
plt.grid()
e = elp(pltdic, 'min',skiplast=0)
plt.gca().add_patch(e[0])
plt.gca().add_patch(e[1])
plt.title("RMSE on CPU usage data",fontsize=15)
plt.savefig("D:/Manrich-PC/Dropbox/Dropbox/M/MThesis/masters/figs/cpu5x_rmse.pdf", type='pdf')
# plt.savefig("D:\Manrich-PC\Dropbox\Dropbox\M\MMondeling\Figs/cpu_rmse.png", type='png')
# Stat test
ma_data = results[0][:,0]
ar_data = results[2][:,0]
calcPvalue(ma_data,ar_data)
```
