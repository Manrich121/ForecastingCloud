```python
%matplotlib inline
import numpy as np
import scipy
from matplotlib import pyplot as plt
import evaluation as eval
from sklearn.metrics import explained_variance_score

from pybrain.datasets import SupervisedDataSet
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.structure.modules import SigmoidLayer, TanhLayer
```

```python
def sampleGeometrically(A, B):
    if A<B:
        logA, logB = scipy.log(A), scipy.log(B)
    else:
        logA, logB = scipy.log(B), scipy.log(A)
    return scipy.exp(scipy.random.uniform(low=logA, high=logB))
```

```python
def randomSearch(trainds, validds, input_size, hidden_range=[4, 32], eta_range=[0.0001, 10.0], activation_func_prob=[.5, .5], lamda_range=[1e-5, 1e-1], epochs_factor=2, miniters=10, maxiters=100):
    bestparams = []
    bestrmse = np.inf
    bestvar = np.inf
    bestiter = maxiters
    bestnet = None
    for iter in range(1, maxiters-1):
#         Sample random hyperparams
        hidden = np.int(sampleGeometrically(hidden_range[0], hidden_range[1])) 
        eta = sampleGeometrically(eta_range[0], eta_range[1])
        lmda = sampleGeometrically(lamda_range[0], lamda_range[1])
        epochs = 20
        print 'Iter:', iter, 'Epochs:', epochs, 'Hidden_size:', hidden, 'Eta:', eta, 'Lamda:', lmda
           
        net = buildNetwork(input_size, hidden, 1, bias=True)
        trainer = BackpropTrainer(net, trainds, learningrate=eta, weightdecay=lmda, shuffle=False)
        trainer.trainEpochs(epochs)
        pred = np.nan_to_num(net.activateOnDataset(validds))
        valid_rmse = eval.calc_RMSE(validds['target'], pred, print_=True)  
        valid_var = np.var(pred) / np.var(validds['target'])
        print 'Variance score:', valid_var
                
        if valid_rmse < bestrmse and np.abs(1 - valid_var) < np.abs(1-bestvar):
            bestrmse = valid_rmse
            bestvar = valid_var
            bestiter = iter
            bestparams = [{'hidden_size':hidden}, {'eta':eta}, {'lamda':lmda}]
            bestnet = net
            
            plt.figure()
            plt.plot(validds['target'])
            plt.plot(pred)
            plt.title(iter)
            
        if bestiter < iter/2 and iter>miniters:
            return bestparams, bestrmse, bestnet
    return bestparams, bestrmse, bestnet
```

```python
data = np.genfromtxt("d:/data/cpuRate/cpu_905814.csv",skip_header=1, delimiter=',',usecols=(1))

TRAIN = 2000
VALID = 200
TEST = 200
INPUT_SIZE = 30

train, valid, test = data[:TRAIN], data[TRAIN-INPUT_SIZE:TRAIN+VALID], data[TRAIN-INPUT_SIZE+VALID:TRAIN+VALID+TEST]

trainds = SupervisedDataSet(INPUT_SIZE, 1)
testds = SupervisedDataSet(INPUT_SIZE, 1)
validds = SupervisedDataSet(INPUT_SIZE, 1)

for i in range(INPUT_SIZE,train.shape[0]):
    trainds.appendLinked(train[i-INPUT_SIZE:i],train[i])

for i in range(INPUT_SIZE,test.shape[0]):
    testds.appendLinked(test[i-INPUT_SIZE:i],test[i])

for i in range(INPUT_SIZE,valid.shape[0]):
    validds.appendLinked(valid[i-INPUT_SIZE:i],valid[i])

hyperparams, rmse, net = randomSearch(trainds, validds, input_size=INPUT_SIZE)
print rmse
print hyperparams

plt.figure()
pred = net.activateOnDataset(testds)
plt.plot(testds['target'])
plt.plot(pred)
plt.title('Test data')
```
