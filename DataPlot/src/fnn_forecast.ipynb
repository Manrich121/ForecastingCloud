```python
%matplotlib inline
import numpy as np
import scipy
import csv
from matplotlib import pyplot as plt
import Fnn_model
from pybrain.tools.customxml.networkreader import NetworkReader
from pybrain.datasets import SupervisedDataSet
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
import evaluation as eval
```

```python
hyperparams = np.genfromtxt("..\data\cpu_networks\hyperparams.csv", delimiter=',', dtype=None, skip_header=1)
print hyperparams[0]
```

```python
curMachine = hyperparams[0][0]
curMachine = curMachine.strip("'")
curEta = hyperparams[0][3]
curLmda = hyperparams[0][4]
print curMachine, curEta, curLmda
```

```python
data = np.genfromtxt("d:/data/cpuRate/"+curMachine+".csv", delimiter=',', skip_header=1, usecols=(1))
plt.plot(data)
```

```python
curNet = NetworkReader.readFrom('../data/cpu_networks/'+curMachine+".xml")
print curNet
```

```python
INPUT_SIZE = 30
str_train = 1000
end_train = 3000
epochs = 20
trainds = SupervisedDataSet(INPUT_SIZE, 1)
for i in range(str_train+INPUT_SIZE, end_train):
        trainds.appendLinked(data[i-INPUT_SIZE:i],data[i])
```

```python
pred = curNet.activateOnDataset(trainds)
plt.figure(figsize=(12,8))
plt.plot(trainds['target'])
plt.plot(pred)
eval.calc_RMSE(trainds['target'], pred, print_=True)
```

```python
unknown = data[3030:3060]
forecasts = []
for i in range(30):
    fc = curNet.activate(unknown)
    forecasts.append(fc)
    unknown = np.append(unknown[1:], fc)
        
plt.plot(data[3030:3060])
plt.plot(forecasts)
```

```python
trainer = BackpropTrainer(curNet, trainds, learningrate=curEta, weightdecay=curLmda, momentum=0.1, shuffle=False, verbose=True)
for i in range(epochs):
    trainer.train()
```

```python
pred = curNet.activateOnDataset(trainds)
plt.figure(figsize=(12,8))
plt.plot(trainds['target'])
plt.plot(pred)
rmse = eval.calc_RMSE(trainds['target'], pred, print_=True)
print rmse*rmse
```
